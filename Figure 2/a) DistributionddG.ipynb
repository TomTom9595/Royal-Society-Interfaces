```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity
from sklearn.utils.fixes import parse_version

plt.rcParams['axes.linewidth'] = 5

y_d = np.linspace(-20, -2, 1000)
y = sorted(y)
y=np.array(y)

from sklearn.neighbors import KernelDensity

# instantiate and fit the KDE model
kde = KernelDensity(bandwidth=1, kernel='epanechnikov')
kde.fit(y[:, None])

# score_samples returns the log of the probability density
logprob2 = kde.score_samples(y_d[:, None])



plt.fill_between(y_d, np.exp(logprob2), alpha=0.5, color='orange', label='300,000\nstrands')
plt.xlabel('$\Delta\Delta$G ($kcal\, mole^{-1}$) ', fontsize=65)
plt.ylabel('Distribution', fontsize=65)
#plt.ylim(0,0.4)
plt.yticks([0.0,0.1,0.2,0.3,0.4])
plt.legend(fontsize=65, loc=2, frameon = False)
plt.axis([-22,0,0,0.45])
plt.xticks([-20, -15, -10, -8.81, -5, 0], ['-20', '-15', '-10', '', '-5', '0'])
#plt.title('$\Delta\Delta$G Distribution', fontsize=55)
plt.tick_params(axis='both', length=15, width=5)
plt.subplots_adjust(bottom=0.2)
plt.subplots_adjust(left=0.2)
plt.savefig("Energy_Distribution.svg",dpi=300, format='svg')
plt.show()
```
