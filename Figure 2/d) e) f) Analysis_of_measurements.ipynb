```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in Excel file
df = pd.read_excel('Measurements_Different_RPs.xlsx')

# Extract time and data columns
time = df.iloc[:, 0]
data = df.iloc[:, 1:]

# Plot time against data
fig, ax = plt.subplots()


#for column in data.columns[4:]:
#    ax.plot(time/3600, data[column], label=column, linewidth=5)
    
ax.plot((time+60)/3600, data[data.columns[0]], label=data.columns[0], linewidth=10)
ax.plot((time+60)/3600, data[data.columns[1]], label=data.columns[1], linewidth=10)
ax.plot((time+60)/3600, data[data.columns[2]], label=data.columns[2], linewidth=10)
ax.plot((time+60)/3600, data[data.columns[3]], label=data.columns[3], linewidth=10)

ax.set_xlim([-0.5, 5])
ax.set_ylim([0, 240])

ax.set_xlabel('Time (h)', fontsize=65)
ax.set_ylabel('Intensity (n.u.)', fontsize=65)
plt.legend(fontsize=55, loc=0, frameon = False)
plt.tick_params(axis='both', length=15, width=5)
plt.subplots_adjust(bottom=0.11)
plt.subplots_adjust(left=0.15)
plt.savefig('Compare N25.svg', format='svg', dpi=300)
plt.show()
```
