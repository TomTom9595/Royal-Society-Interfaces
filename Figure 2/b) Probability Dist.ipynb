```python
import matplotlib.pyplot as plt
import numpy as np
kB = 1.9872*pow(10,-3)
T = 302.15
kBT = kB*T

plt.ion()
plt.rcParams['xtick.labelsize']=65
plt.rcParams['ytick.labelsize']=65
plt.rcParams['ytick.right'] = plt.rcParams['ytick.labelright'] = False
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = True
plt.rcParams.update({'figure.autolayout': False})
plt.rcParams['figure.figsize'] = 21, 15
plt.rcParams['axes.linewidth'] = 5

plt.rcParams['font.family'] = 'Arial'

Su20N50EnergiesList = []
Su20N50Energies = open("N25_Energies_sorted.txt")
for line in Su20N50Energies:
    Su20N50EnergiesList.append(line.rstrip())
Su20N50Energies.close()
x = [float(i) for i in Su20N50EnergiesList]

probs=[]


    
bins = np.linspace(min(x), max(x), 100)

for i in bins:
    probs.append(np.exp(-i)/kBT)

problist=[]

for i in probs:
    problist.append(i/sum(probs))

#print(sum(problist[:30]))
plt.plot(bins,problist, color='tab:blue', linewidth=20, label='Random Pool')
plt.xlabel('$\Delta\Delta$G ($kcal\, mole^{-1}$) ', fontsize=65)
plt.ylabel('Binding probability', fontsize=65)
plt.axis([-19.5,0, -0.005, 0.16])
plt.xticks([-20, -15, -10, -8.81, -5, 0], ['-20', '-15', '-10', '', '-5', '0'])
plt.tick_params(axis='both', length=15, width=5)
plt.legend(fontsize=65, frameon = False, loc=1, markerscale=1.)
plt.subplots_adjust(bottom=0.2)
plt.subplots_adjust(left=0.2)
plt.savefig("ProbabilityDist.svg",dpi=900, format='svg')
plt.show()
```
